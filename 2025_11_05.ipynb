```python
word: str = "banana"

def char_frequency(word: str) -> dict[str, int]:
    char_freqs: dict[str, int] = {}

    for char in word.lower().replace(" ", ""):
        if char not in char_freqs:
            char_freqs[char] = 1
        else:
            char_freqs[char] += 1

    return char_freqs

char_frequency(word)
```

```python
numbers: list[int] = [3, 1, 2, 3, 2, 4]

def unique_numbers(numbers: list[int]) -> list[int]:
    seen_nums: list[int] = []

    for num in numbers:
        if num not in seen_nums:
            seen_nums.append(num)

    return seen_nums

unique_numbers(numbers)
```

```python
sentence: str = "The quick brown fox jumps over the lazy dog"

def longest_word(sentence: str) -> tuple[str, int]:
    sentence.replace(".", "").replace(",", "").replace("!", "").replace("?", "")

    words = sentence.split()

    longest = ""
    max_len = 0

    for word in words:
        if len(word) > max_len:
            longest = word
            max_len = len(word)

    return (longest, max_len)
 
longest_word(sentence)
```
