```python
word: str = "banana"

def char_frequency(word: str) -> dict[str, int]:
    char_freqs: dict[str, int] = {}

    for char in word.lower().replace(" ", ""):
        if char not in char_freqs:
            char_freqs[char] = 1
        else:
            char_freqs[char] += 1

    return char_freqs

char_frequency(word)
```
